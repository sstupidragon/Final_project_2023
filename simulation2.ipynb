```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
x1=np.linspace(0,0.99,10)
x2=np.linspace(1.01,10.01,100)
sigma2=1

def y1(x,r):
    return sigma2*x/(1-x)
def y2(x,r):
    return r*(x-1)/x+1/(x-1)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
for r in [1,2.33,3.66,5]:
    ax.plot(np.append(x1,x2),np.append(y1(x1,r),y2(x2,r)))
ax.set_ylim(0,10)
```

Generate_data

```python
def generate_data(n,p,SIGMA,sigma,beta):    
    x=np.random.multivariate_normal([0 for i in range(p)], SIGMA, size=n)
    e=np.random.normal(loc=0, scale=sigma, size=n)
    y=beta@x+e
    return x,y,e
```
